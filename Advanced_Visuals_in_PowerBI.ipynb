# Question 1. What is the difference between a Basic Filter, Advanced Filter, and a Slicer in Power BI?     
    
In Power BI, Basic Filters, Advanced Filters, and Slicers all help you refine the data displayed in your reports, but they offer different levels of control and interaction:

- **Basic Filter**: This is the simplest type of filter. It allows you to select one or more values from a list to include or exclude data. Think of it like a checklist where you tick off the items you want to see.

- **Advanced Filter**: This provides more sophisticated filtering capabilities than a basic filter. It allows you to use operators like "contains," "starts with," "is greater than," "is less than," "is not blank," etc., to define custom conditions. You can also combine multiple conditions using "AND" or "OR" logic.

- **Slicer**: A Slicer is a visual, interactive filter that allows users to directly select values on the report canvas to filter the data. Slicers are user-friendly and provide immediate visual feedback. They can be configured to allow single selection, multi-selection, or dropdowns, and they work across all visuals on the same report page (or across multiple pages if synchronized).

# Question 2.  Using the Sales dataset, apply a Basic Filter to display only the sales data for the year 2023.  

First, let's create a dummy sales dataset to demonstrate the filtering process.

```python
import pandas as pd
import numpy as np

# Create a sample Sales DataFrame
dates = pd.to_datetime(pd.date_range(start='2022-01-01', end='2023-12-31', freq='D'))
sales_data = {
    'OrderDate': dates,
    'Region': np.random.choice(['East', 'West', 'North', 'South'], len(dates)),
    'Product': np.random.choice(['A', 'B', 'C', 'D'], len(dates)),
    'Sales': np.random.randint(100, 1000, len(dates))
}
df_sales = pd.DataFrame(sales_data)

print("Original Sales Data (first 5 rows):")
display(df_sales.head())
```

Now, let's apply a Basic Filter to display only the sales data for the year 2023.

```python
# Ensure 'OrderDate' is in datetime format (if not already)
df_sales['OrderDate'] = pd.to_datetime(df_sales['OrderDate'])

# Apply the basic filter for the year 2023
sales_2023 = df_sales[df_sales['OrderDate'].dt.year == 2023]

print("Sales Data for the Year 2023 (first 5 rows):")
display(sales_2023.head())
print(f"Total sales records for 2023: {len(sales_2023)}")
print(f"Earliest date in 2023 sales: {sales_2023['OrderDate'].min()}")
print(f"Latest date in 2023 sales: {sales_2023['OrderDate'].max()}")
```

# Question 3. Create a Pie Chart of Sales by Product and add Slicer that allows users to filter the dataset by Region.

First, let's create a Pie Chart showing the overall sales distribution by product.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate total sales by product
sales_by_product = df_sales.groupby('Product')['Sales'].sum().reset_index()

# Create the pie chart
fig1, ax1 = plt.subplots(figsize=(8, 8))
ax1.pie(sales_by_product['Sales'], labels=sales_by_product['Product'], autopct='%1.1f%%', startangle=90, colors=sns.color_palette('viridis', len(sales_by_product)))
ax1.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Overall Sales Distribution by Product')
plt.show()
```

Now, let's simulate the 'slicer' functionality. Imagine a user selects 'East' from a Region slicer. We will filter the data for the 'East' region and display a new pie chart.

```python
# Simulate a slicer for 'Region' (e.g., selecting 'East')
filtered_sales_east = df_sales[df_sales['Region'] == 'East']

# Calculate sales by product for the 'East' region
sales_by_product_east = filtered_sales_east.groupby('Product')['Sales'].sum().reset_index()

# Create a pie chart for the 'East' region
fig2, ax2 = plt.subplots(figsize=(8, 8))
ax2.pie(sales_by_product_east['Sales'], labels=sales_by_product_east['Product'], autopct='%1.1f%%', startangle=90, colors=sns.color_palette('viridis', len(sales_by_product_east)))
ax2.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Sales Distribution by Product for East Region (Simulated Slicer)')
plt.show()
```

#  Question 4. What is a Visual-Level Filter, and how does it differ from Page-Level and Report-Level Filters?



 # In Power BI, filters can be applied at different levels, each affecting the data visibility in a specific scope:

- **Visual-Level Filter**: This filter applies only to a single visual (e.g., a specific chart, table, or map) on a report page. It allows you to refine the data displayed within that individual visual without affecting other visuals on the same page or throughout the report. For example, you might have a bar chart showing sales by product, and you apply a visual-level filter to show only products with sales greater than $1000 on that specific chart.

- **Page-Level Filter**: This filter applies to all visuals on a specific report page. When you set a page-level filter, every visual on that page will display data refined by that filter, but visuals on other pages of the report will remain unaffected. For instance, if you have a page dedicated to 'North Region' sales, you would apply a page-level filter for Region = 'North'.

- **Report-Level Filter**: This filter is the broadest and applies to all visuals across all pages of the entire Power BI report. Any data shown anywhere in the report will adhere to the conditions set by report-level filters. An example would be applying a report-level filter for Year = 2023 to ensure all data displayed in the entire report is only for the year 2023.

# Question 6. What is the role of Drill-Down in Power BI? How does it improve data exploration?    
   
In Power BI, Drill-Down is a powerful feature that allows users to explore data at different levels of granularity within a visual. It enables you to move from a higher-level summary of data to more detailed information by clicking on a data point.

Here's how it works and how it improves data exploration:

- **Hierarchical Exploration**: Drill-down typically works with data hierarchies (e.g., Year -> Quarter -> Month -> Day, or Country -> State -> City). When you drill down, you move to the next level of the hierarchy, revealing more granular data for the selected category.

- **Enhanced Detail**: Instead of having multiple visuals for different levels of detail, drill-down allows a single visual to serve multiple purposes. For example, a bar chart showing sales by year can be drilled down to show sales by quarter, then by month, for a specific year, all within the same visual.

- **Improved Data Exploration:**

- **Contextual Analysis**: It helps users understand the context behind aggregated numbers. If total sales are high, drilling down can show which products, regions, or time periods contributed most.      

- **Root Cause Analysis**: When a trend or anomaly is observed at a high level, drill-down can quickly help identify the specific categories or sub-categories driving that observation.    

- **Interactive Discovery**: It makes data exploration more interactive and intuitive, allowing users to follow their line of inquiry without having to create new reports or visuals.   

- **Space Efficiency**: It conserves report canvas space by allowing one visual to display various levels of detail, making reports cleaner and easier to digest.

# Question 7. Using the dataset, create a hierarchy drill-down from    Country → Region → Product Category to analyze sales at different levels.
  

First, let's visualize the highest level of our hierarchy: **Sales by Country**.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate total sales by Country
sales_by_country = df_sales.groupby('Country')['Sales'].sum().reset_index()

# Create a bar chart for sales by Country
plt.figure(figsize=(10, 6))
sns.barplot(x='Country', y='Sales', data=sales_by_country, palette='viridis')
plt.title('Total Sales by Country (Drill-Down Level 1)')
plt.xlabel('Country')
plt.ylabel('Total Sales')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Now, let's simulate a drill-down into a specific country, for example, **'India'**. We will then see the sales distribution by **Region** within India.

```python
# Filter data for 'India' (simulating drill-down)
sales_india = df_sales[df_sales['Country'] == 'India']

# Calculate sales by Region within India
sales_india_by_region = sales_india.groupby('Region')['Sales'].sum().reset_index()

# Create a bar chart for sales by Region in India
plt.figure(figsize=(10, 6))
sns.barplot(x='Region', y='Sales', data=sales_india_by_region, palette='magma')
plt.title('Total Sales by Region in India (Drill-Down Level 2)')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Finally, let's drill down further into a specific region within India, for instance, the **'East'** region. We will then analyze sales by **Product Category** (our 'Product' column) within 'India' and 'East' region.

```python
# Filter data for 'India' and 'East' region (simulating drill-down)
sales_india_east = df_sales[(df_sales['Country'] == 'India') & (df_sales['Region'] == 'East')]

# Calculate sales by Product within India and East region
sales_india_east_by_product = sales_india_east.groupby('Product')['Sales'].sum().reset_index()

# Create a bar chart for sales by Product in India's East region
plt.figure(figsize=(10, 6))
sns.barplot(x='Product', y='Sales', data=sales_india_east_by_product, palette='rocket')
plt.title('Total Sales by Product in India - East Region (Drill-Down Level 3)')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# Question 8. Implement Drill-Through so that selecting a Product navigates to a detailed report page showing regionwise sales.  

In a Power BI report, Drill-Through allows users to right-click on a data point (e.g., a specific product in a chart) and navigate to a pre-defined 'drill-through page' that shows more detailed information, with filters applied based on the original selection.

Here, we'll simulate this by creating a Python function. This function will act as our 'detailed report page' and will dynamically filter and visualize data based on a selected 'Product'. We'll use our `df_sales_india` DataFrame to maintain the report-level filter for India.

```python
import matplotlib.pyplot as plt
import seaborn as sns

def show_product_region_sales(product_name):
    """
    Simulates a drill-through report page showing region-wise sales for a selected product.
    """
    print(f"\n--- Drill-Through: Detailed Sales Report for Product '{product_name}' ---")

    # Filter the 'report-level' India sales data for the selected product
    product_sales_data = df_sales_india[df_sales_india['Product'] == product_name]

    if product_sales_data.empty:
        print(f"No sales data found for Product '{product_name}' in India.")
        return

    # Calculate sales by Region for the selected product
    sales_by_region_for_product = product_sales_data.groupby('Region')['Sales'].sum().reset_index()

    # Create a bar chart for region-wise sales of the selected product
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Region', y='Sales', data=sales_by_region_for_product, palette='coolwarm')
    plt.title(f'Region-wise Sales for Product "{product_name}" in India')
    plt.xlabel('Region')
    plt.ylabel('Total Sales')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# --- Demonstration of Drill-Through ---
```

Let's assume a user 'drills through' from a visual by selecting 'Product B'. We would then call our simulation function with 'Product B'.

```python
# Get a list of unique products from our India sales data to pick from
available_products = df_sales_india['Product'].unique()

if len(available_products) > 0:
    # Select a product for demonstration (e.g., the first one available or a specific one)
    selected_product = available_products[0] # or choose 'B', 'C', etc. if you know they exist
    print(f"Simulating drill-through for selected product: '{selected_product}'")
    show_product_region_sales(selected_product)
else:
    print("No products available in the filtered India sales data to demonstrate drill-through.")
```

We can also try another product, for example 'Product C', if it exists in our dataset.

```python
if 'C' in available_products:
    print("Simulating drill-through for selected product: 'C'")
    show_product_region_sales('C')
else:
    print("Product 'C' not found in the filtered India sales data.")
```
